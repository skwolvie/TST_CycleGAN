```python
df['formal_1']= df['formal_1'].apply(lambda x: str.encode(x, encoding='UTF-8').decode('utf-8'))
df['informal_0']= df['informal_0'].apply(lambda x: str.encode(x, encoding='UTF-8').decode('utf-8'))
```

```python
for index, row in df.iterrows():
    with open('data.txt', 'a', encoding='utf-8') as f:
        f.writelines('0 +++$+++ '+ str(row['informal_0'])+ "\n")
        f.writelines('1 +++$+++ '+ str(row['formal_1'])+ "\n")

f.close()
```

```python
# create dictionary
import gensim
from gensim import corpora
from gensim.utils import simple_preprocess
gensim_dictionary = corpora.Dictionary(simple_preprocess(line, deacc =False) for line in open('data.txt', encoding='utf-8'))

print("The dictionary has: " +str(len(gensim_dictionary)) + " tokens")
for k, v in gensim_dictionary.token2id.items():
     print(f'{k:{15}} {v:{10}}')

for k, v in gensim_dictionary.token2id.items():
    with open('dict.txt', 'a', encoding='utf-8') as f:
        f.writelines(k+" "+str(v)+'\n')
```

```python
with open('data.txt', 'r', encoding='UTF-8') as file:
    lines = file.readlines()
    lines = [line.rstrip() for line in lines]

print(len(lines))
print(lines[0])
```

```python
text_tokens = [[text for text in line.split()] for line in lines]
text_tokens[:10]
dict_LoS = corpora.Dictionary(text_tokens)
print(dict_LoS)
print(i for i in dict_LoS.token2id)
```
